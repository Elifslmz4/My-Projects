```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

file_path = 'Ecol_Raman.csv'
data = pd.read_csv(file_path)

components = [
    'protein', 'rna', 'gsh', 'dna', 'atp', 'asp', 'gln', 'utp', 'gtp',
    'glucose', 'val', 'fru-bp-1,4', 'glu', 'lipid', 'NAD+'
]

for component in components:
    plt.figure()
    plt.plot(data['wavenumber'], data[component], label=component)
    plt.xlabel('Wavenumber')
    plt.ylabel(component)
    plt.title(f'Wavenumber vs {component}')
    plt.legend()
    plt.grid(True)
    plt.show()
```

```python
plt.figure(figsize=(12, 8))

# Plot each component
for column in data.columns[:-1]:  # Exclude 'wavenumber'
    plt.plot(data['wavenumber'], data[column], label=column, alpha=0.7)

plt.xlabel('Wavenumber')
plt.ylabel('Intensity')
plt.title('Raman Spectroscopy: Wavenumber vs Various Components')
plt.legend(loc='upper right', bbox_to_anchor=(1.15, 1))
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# Load one of the processed mean spectra files (e.g., stationary phase as an example)
raw_df = pd.read_csv("Ecol_Raman_ec_sta_mean_raman.csv")

# Simulate a simple baseline (linear trend)
baseline = 0.005 * raw_df['Wavenumber'] - 5
corrected_intensity = raw_df['Intensity'] - baseline

# Plot raw vs. baseline-corrected spectrum
plt.figure(figsize=(12, 6))
plt.plot(raw_df['Wavenumber'], raw_df['Intensity'], label='Raw Spectrum', color='red')
plt.plot(raw_df['Wavenumber'], corrected_intensity, label='Baseline-Corrected Spectrum', color='blue')
plt.plot(raw_df['Wavenumber'], baseline, label='Estimated Baseline', linestyle='--', color='gray')

plt.xlabel("Raman Shift (cm⁻¹)")
plt.ylabel("Intensity (a.u.)")
plt.title("Raw vs. Baseline-Corrected Raman Spectrum")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Load the data
file_path = 'Ecol_Raman.csv'
data = pd.read_csv(file_path)

components = [
    'protein', 'rna', 'gsh', 'dna', 'atp', 'asp', 'gln', 'utp', 'gtp',
    'glucose', 'val', 'fru-bp-1,4', 'glu', 'lipid', 'NAD+'
]

plt.figure(figsize=(12, 8))

offset = 0
offset_step = 1.2  # adjust to control spacing between lines

for component in components:
    plt.plot(data['wavenumber'], data[component] + offset, label=component)
    offset += offset_step  # increase offset for next component

plt.xlabel('Wavenumber (cm$^{-1}$)')
plt.ylabel('Intensity (offset for clarity)')
plt.title('Waterfall Plot of Reference Raman Spectra')
plt.legend(loc='center left', bbox_to_anchor=(1, 0.5))
plt.grid(True)
plt.tight_layout()
plt.show()
```
